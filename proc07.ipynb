```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from dezero import Variable
```

```python
def rosebrock(x0, x1):
    return 100 * (x1 - x0 ** 2) ** 2 + (1 - x0) ** 2
```

```python
x0 = Variable(np.array(0.0))
x1 = Variable(np.array(2.0))
```

```python
y = rosebrock(x0, x1)
```

```python
y
```

```python
y.backward()
```

```python
x0.grad, x1.grad
```

```python
x0 = Variable(np.array(0.))
x1 = Variable(np.array(2.0))
lr = 1e-3
epochs = 1000
x0s = np.array([x0.data])
x1s = np.array([x1.data])
for _epoch in range(epochs):
    print(f"x0={x0}, x1={x1}")
    x0.cleargrad()
    x1.cleargrad()
    x0.creator = None
    x1.creator = None
    y = rosebrock(x0, x1)
    y.backward()
    x0 -= lr * x0.grad
    x1 -= lr * x1.grad
    x0s = np.append(x0s, x0.data)
    x1s = np.append(x1s, x1.data)
plt.plot(x0s, x1s)
plt.show()
```

```python
def f(x):
    return x**4 - 2 * x**2


def gx2(x):
    return 12 * x**2 - 4
```

```python
x = Variable(np.array(2.))
iters = 10
```

```python
for i in range(iters):
    print(i, x)
    x.cleargrad()
    x.creator = None
    y = f(x)
    y.backward()
    x -= x.grad / gx2(x)
```

```python
from dezero import Function
class Sin(Function):
    def forward(self, x):
        return np.sin(x)

    def backward(self, gy):
        x = self.inputs[0].data
        return gy * np.cos(x)
```

```python
def sin(x):
    return Sin()(x)
```

```python
x = Variable(np.array(1.))
y = sin(x)
```
